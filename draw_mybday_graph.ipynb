# 내 생일 그래프 그리기
## 19990211 송병준
* 내가 태어난 년도부터 매년 내 생일의 최고 기온과 최저 기온의 그래프 그리기

```python
import csv

f = open('incheon_weather.csv', 'r')
data = csv.reader(f)
next(data)  # CSV 컬럼 헤더 ㅃ2

years = []
highs = []
lows = []

for row in data:
    try:
        year, month, day_of_month = map(int, row[0].split('-'))
        low, high = map(float, row[3:])
    except ValueError:
        continue

    if year >= 1999 and month == 2 and day_of_month == 11:
        years.append(year)
        highs.append(high)
        lows.append(low)

f.close()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumBarunGothic' # 폰트 깨지면 이 라인 지워주세요~
plt.rcParams['axes.unicode_minus'] = False

plt.plot(highs, color='red', linestyle='dashed', marker='v', label='최고 기온')
plt.plot(lows, color='blue', linestyle='dashed', marker='o', label='최저 기온')

plt.title("내 생일 최고 & 최저 기온 변화 그래프")
plt.xlabel("연도")
plt.ylabel("기온")
plt.xticks(ticks=range(len(years)), labels=years, rotation=45)
plt.legend()

plt.show()
```
